##### Features: sepal length, sepal width, petal length, petal width (all in cm)
##### Class: Iris Setosa, Iris Versicolour and Iris Virginica


### 1. Loading the relevant libraries

```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import warnings
import os
warnings.filterwarnings("ignore")
```

### 2. Data loading

```python
iris = load_iris()
data = pd.DataFrame(iris.data, columns=iris.feature_names)
data['species'] = iris.target
data['species'] = data['species'].map({0: 'Setosa', 1: 'Versicolour', 2: 'Virginica'})
```

### 5. Relationships between features

#### Scatter plots for feature relationships

```python
colors=['#FFA500','#9B59B6','#008080']
species=data['species'].unique()
```

```python
for i in range(3):
    x=data[data['species']==species[i]]
    plt.scatter(x['sepal length (cm)'],x['sepal width (cm)'],c= colors[i],label=species[i])

plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.title('Sepal length vs Sepal width')
plt.legend()
```

```python
for i in range(3):
    x=data[data['species']==species[i]]
    plt.scatter(x['petal length (cm)'],x['petal width (cm)'],c= colors[i],label=species[i])

plt.xlabel("Petal length")
plt.ylabel("Petal width")
plt.title('Petal length vs Petal width')
plt.legend()
```

```python
for i in range(3):
    x=data[data['species']==species[i]]
    plt.scatter(x['sepal length (cm)'],x['petal length (cm)'],c= colors[i],label=species[i])

plt.xlabel('Sepal length')
plt.ylabel('Petal length')
plt.title('Sepal length vs Petal length')
plt.legend()
```

```python
for i in range(3):
    x=data[data['species']==species[i]]
    plt.scatter(x['sepal width (cm)'],x['petal width (cm)'],c= colors[i],label=species[i])

plt.xlabel('Sepal width')
plt.ylabel('Petal width')
plt.title('Sepal width vs Petal width')
plt.legend()
```

### 6. Exploring feature interactions and patterns

#### Violin plots for feature distributions by species

```python
palette = "tab10"
# Create a 2x2 subplot for the violin plots
plt.figure(figsize=(15, 10))

# Violin plot for petal length
plt.subplot(2, 2, 1)
sns.violinplot(x='species', y='petal length (cm)', data=data, palette=palette)
plt.title('Petal length by species')

# Violin plot for petal width
plt.subplot(2, 2, 2)
sns.violinplot(x='species', y='petal width (cm)', data=data, palette=palette)
plt.title('Petal width by species')

# Violin plot for sepal length
plt.subplot(2, 2, 3)
sns.violinplot(x='species', y='sepal length (cm)', data=data, palette=palette)
plt.title('Sepal length by species')

# Violin plot for sepal width
plt.subplot(2, 2, 4)
sns.violinplot(x='species', y='sepal width (cm)', data=data, palette=palette)
plt.title('Sepal width by species')

plt.tight_layout()
plt.show()
```

#### Pair plot

```python
sns.pairplot(data, hue="species", size=3)
```

#### Heatmap

```python
plt.figure(figsize=(10,8))
sns.heatmap(data.corr(numeric_only=True),annot=True,cmap='coolwarm')
```
